```python
# Calcular Area de un triangulo
# Variables A, b, h
# Formula A = (b * h) / 2
# Capturar el valor de b y h
import math

print("Calcular Area de un triangulo")

b = float(input("Ingrese la base del triangulo: "))
h = float(input("Ingrese la altura del triangulo: "))

A = (b * h) / 2

print("El area del triangulo es: ", A)
print("Programa realizado por Jesus Arriaga NL = 5")
```
